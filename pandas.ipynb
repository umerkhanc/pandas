```python
# Import the necessary libraries
import pandas as pd
import numpy as np

# Create a DataFrame from a dictionary
data = {'name': ['John', 'Jane', 'Sam', 'Sara'],
        'age': [32, 28, 45, 55],
        'city': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df)

# Create a DataFrame from a list of lists
data = [['John', 32, 'New York'],
        ['Jane', 28, 'San Francisco'],
        ['Sam', 45, 'Los Angeles'],
        ['Sara', 55, 'Chicago']]
df = pd.DataFrame(data, columns=['name', 'age', 'city'])
print(df)

# Create a DataFrame from a CSV file
df = pd.read_csv('my_data.csv')
print(df)

# Get the first n rows of a DataFrame
print(df.head(2))

# Get the last n rows of a DataFrame
print(df.tail(2))

# Get the columns of a DataFrame
print(df.columns)

# Get the data types of the columns
print(df.dtypes)

# Get summary statistics of the DataFrame
print(df.describe())

# Get the values of a DataFrame as a numpy array
print(df.values)

# Get the value of a specific cell
print(df.at[0,'name'])

# Get the value of a specific cell
print(df.iat[0,0])

# Get a slice of the DataFrame
print(df[1:3])

# Get a specific column
print(df['name'])

# Get multiple columns
print(df[['name', 'city']])

# Get a specific row
print(df.loc[1])

# Get multiple rows
print(df.loc[[1,3]])

# Get a specific value
print(df.loc[1,'name'])

# Get multiple values
print(df.loc[1,['name','city']])

# Get a specific value using index and column labels
print(df.at[1,'name'])

# Get a specific value using index and column positions
print(df.iat[1,0])

# Sort the DataFrame by a specific column
print(df.sort_values(by='age'))

# Sort the DataFrame by multiple columns
print(df.sort_values(by=['age','name']))

# Group the DataFrame by a specific column
print(df.groupby('city').mean())
```
